```python
"""
ライブラリpandasを使ったcsvデータの読み込み,sepで区切り記号を設定,もし日本語の場合,encodingを設定する必要があり
事前準備（必要なパッケージをインストール）
pip install pandas
pip install scikit-learn 
pip install matplotlib 
"""
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

# 身長と体重データの読み込み
df = pd.read_csv('height_weight_data.csv',sep=',',encoding="utf-8")
df.head()
```

```python
#特定の列だけ抽出(身長)
df.loc[:, ['height_cm']]
```

```python
# 身長と体重データの平均値算出 mean
mean_values = df[['height_cm','weight_kg']].mean()
print("平均値")
print(mean_values)
```

```python
# グラフ化
plt.figure(figsize=(8, 5))

# scatterで点表示
plt.scatter(df.loc[:, ['height_cm']],df.loc[:, ['weight_kg']], c="blue", marker='>')
plt.xlabel(f'身長(cm)', fontname='MS Gothic')
plt.ylabel(f'体重(kg)', fontname='MS Gothic')
plt.show()
```

```python
# pandas内部提供の機能を利用して、グラフを描画（s: マーカーのサイズ, c: マーカーの色、marker: マーカーの形状）
# cf. https://pygraph.helve-blog.com/pandas/pd-scatter-plot

df.plot.scatter(x="height_cm", y="weight_kg", s=40, c="green", marker="v")
```
